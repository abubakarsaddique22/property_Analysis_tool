```python
df=pd.read_csv('/content/data_version_1.csv')
df.head(10)
```

#### Area column


```python
df['Area']
```

```python
#convert into square foot

# Conversion factors
KANAL_TO_SQFT = 5445
MARLA_TO_SQFT = 272.25
SQYD_TO_SQFT = 9

# Function to convert area to Sq.Ft.
def convert_to_sqft(area):
    if pd.isna(area):  # Handle NaN values
        return np.nan
    area = str(area)  # Ensure it's a string
    if "Kanal" in area:
        num = float(re.findall(r"[\d.]+", area)[0])
        return round(num * KANAL_TO_SQFT)
    elif "Marla" in area:
        num = float(re.findall(r"[\d.]+", area)[0])
        return round(num * MARLA_TO_SQFT)
    elif "Sq. Yd." in area:
        num = float(re.findall(r"[\d.]+", area)[0])
        return round(num * SQYD_TO_SQFT)
    else:
        return None  # If format is unknown

# Apply conversion to the "Area" column
df['area'] = df["Area"].apply(convert_to_sqft)

# Function to ensure no decimal points in output
def format_number(value):
    if pd.isna(value):
        return value
    return int(round(value))  # Always round and convert to integer

# Apply formatting to the "area" and "price_per_sqft" columns
df['area'] = df['area'].apply(format_number)
# df["price_per_sqft"] = df["price_per_sqft"].apply(format_number)
```

```python
#make a column price per sqft
df["price_per_sqft"] = round(df["price"] * 10000000 / df["area"],2)
```

```python
# drop the Area column
df.drop("Area",axis=1,inplace=True)
```

#### subname column

```python

# Extract colony and province from "sub_name" column
def extract_colony_province(sub_name):
    parts = sub_name.split(",")  # Assuming values are separated by commas
    colony = parts[0].strip() if len(parts) > 0 else None
    province = parts[-1].strip() if len(parts) > 1 else None
    return colony, province

# Apply extraction function
df[['colony', 'province']] = df['sub_name'].apply(lambda x: pd.Series(extract_colony_province(x)))
```

```python
df['colony']=df['colony'].str.split('-').str.get(0)
```

```python
df['colony']=df['colony'].str.replace('Al','Al Ahmad')
```
