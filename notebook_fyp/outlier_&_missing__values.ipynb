```python

# Step 1: Identify Outliers using IQR method
Q1 = df["Parking Spaces"].quantile(0.25)
Q3 = df["Parking Spaces"].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove Outliers
df = df[(df["Parking Spaces"] >= lower_bound) & (df["Parking Spaces"] <= upper_bound)]


df['Parking Spaces'].value_counts()
```

```python
correlation = df.corr(numeric_only=True)["Parking Spaces"]["price"]
print(f"Correlation with price: {correlation}")
```

### missing value impute

```python
# Fill missing values with median
df["Parking Spaces"].fillna(df["Parking Spaces"].median(), inplace=True)
```

```python
# Final check for missing values
missing_after_imputation = df["Parking Spaces"].isnull().sum()
print(f"Missing values after imputation: {missing_after_imputation}")
```

### Property Type

-  Fills missing "Property Type" values based on area size
-  Houses for large properties ( 2500 sq ft).
-  Flats for small properties ( 1500 sq ft).
-  Lower for medium properties (1500-2500 sq ft) with high price.
-  Upper for medium properties with lower price.
-  Defaults to "Houses" if area is missing.

```python
df.isnull().sum()
```

```python
df[df['property Type'].isnull()]
```

```python
df[df['property Type'].isnull()][['City','property Type']]
```

```python
# Compute the median price for properties in the medium area range (1500 - 2500 sq ft)
median_price_medium_area = df[(df["area"] > 1500) & (df["area"] <= 2500)]["price"].median()

# Function to impute missing property types based on area and price
def impute_property_type(row):
    if pd.isnull(row["property Type"]):  # If missing
        if row["area"] >= 2500:
            return "Houses"
        elif row["area"] <= 1500:
            return "Flats"
        elif 1500 < row["area"] <= 2500:
            return "Lower" if row["price"] > median_price_medium_area else "Upper"
        else:
            return "Houses"  # Default to Houses if area is also missing
    return row["property Type"]  # Keep existing value

# Apply the function to fill missing values
df["property Type"] = df.apply(impute_property_type, axis=1)
```

```python
print(df["property Type"].value_counts())
```

```python
print(df.isnull().sum())
```

### city

```python
# Most frequent city (for fallback)
most_common_city = df["City"].mode()[0]

# Function to fill missing city values
def impute_city(row):
    if pd.isnull(row["City"]):  # If missing
        if "Lahore" in str(row.values):
            return "Lahore"
        elif "Rawalpindi" in str(row.values):
            return "Rawalpindi"
        elif "Karachi" in str(row.values):
            return "Karachi"
        else:
            return most_common_city  # Fill with most common city
    return row["City"]  # Keep existing value

# Apply the function to fill missing values
df["City"] = df.apply(impute_city, axis=1)
print(df["City"].value_counts())
```
